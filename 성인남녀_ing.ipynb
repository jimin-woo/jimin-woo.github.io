```python
# 전사 텍스트
o = open("2020_일반남녀_학습DB_PCM2TEXT.list", 'r',encoding='cp949') 
k = o.readlines()
adult_text = []
for i in k:
    index = i.index('PCM')
    string = i[index+4 : -1]
    adult_text.append(string)
```

```python
# citycode
o = open("2020_일반남녀_학습DB_PCM2TEXT.list", 'r',encoding='cp949') 
k = o.readlines()
adult_city = []
for i in k:
    index = i.index('00')
    if i[index+35] == '수':
        string = i[index+35:index+38]
        adult_city.append(string)
    else:
        string = i[index+35:index+37]
        adult_city.append(string)
```

```python
# A_001 - setnum2
o = open("2020_일반남녀_학습DB_PCM2TEXT.list", 'r',encoding='cp949') 
k = o.readlines()
adult_setnum2 = []
for i in k:
    index = i.index('00')
    string = i[index+18:index+23]
    adult_setnum2.append(string)
```

```python
# 성인남녀_001 - setnum
o = open("2020_일반남녀_학습DB_PCM2TEXT.list", 'r',encoding='cp949') 
k = o.readlines()
adult_setnum = []
for i in k:
    index = i.index('00')
    string = i[index+9:index+17]
    adult_setnum.append(string)
```

```python
# gender
o = open("2020_일반남녀_학습DB_PCM2TEXT.list", 'r',encoding='cp949') 
k = o.readlines()
adult_gender = []
for i in k:
    index = i.index('00')
    string = i[index+24:index+25]
    adult_gender.append(string)
```

```python
# age
o = open("2020_일반남녀_학습DB_PCM2TEXT.list", 'r',encoding='cp949') 
k = o.readlines()
adult_age = []
for i in k:
    index = i.index('00')
    string = i[index+32:index+34]
    adult_age.append(string)
```

```python
# nameid
o = open("2020_일반남녀_학습DB_PCM2TEXT.list", 'r',encoding='cp949') 
k = o.readlines()
adult_Nameid = []
for i in k:
    index = i.index('00')
    string = i[index+26:index+31]
    adult_Nameid.append(string)
```

```python
# 스크립트 번호
o = open("2020_일반남녀_학습DB_PCM2TEXT.list", 'r',encoding='cp949') 
k = o.readlines()
adult_script = []
for i in k:
    index = i.index('PCM')
    string = i[index-6 : index-1]
    adult_script.append(string)
```

```python
# adult_text -> target 변수
# adult_city 
# adult_setnum2 
# adult_setnum
# adult_gender
# adult_age
# adult_Nameid
# adult_script
```

```python
# 성인남여 information dictionary
adult_dict = {}
adult_dict['Nameid'] = adult_Nameid # 추출한 리스트
adult_dict['age'] = adult_age
adult_dict['city'] = adult_city
adult_dict['gender'] = adult_gender
adult_dict['setnum'] = adult_setnum
adult_dict['setnum2'] = adult_setnum2
adult_dict['scriptnum'] = adult_script
adult_dict['text'] = adult_text

# 성인남여 information DataFrame
df_adult = pd.DataFrame(adult_dict)
df_adult.head()
```
